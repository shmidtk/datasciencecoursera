```python
import numpy
import timeit

#создаем массивы
list1 = list(range(10000000))

list2 = numpy.arange(10000000, dtype=numpy.int64)

sum1 = 0
sum2 = 0
```

```python
# Замеряем время для цикла for
%%timeit -n1 -r1

sum1=0
for x in list1: 
    sum1+=x
```

```python
%%timeit -n1 -r1
sum2=0
for x in list2:
    sum2 = sum2 + x
```

```python
# Замеряем время для фукнции sum
%%timeit -n1 -r1
sum(list1)
```

```python
%%timeit -n1 -r1
sum(list2)
```

```python
# Замеряем время для фукнции numpy.sum
%%timeit -n1 -r1
numpy.sum(list1)
```

```python
%%timeit -n1 -r1
numpy.sum(list2)
```

Функция numpy.sum написана на С, а так же использует векторные процессорные вычисления. Это ускоряет процесс обращения к памяти и оптимизирует загрузку процессора, вычисления происходят сразу по многих элементам складываемых векторов, а не по отдельным элементам.
